```python
from sklearn.preprocessing import LabelEncoder

# Membuat objek LabelEncoder
label_encoder = LabelEncoder()

# Mengubah kolom 'Gender' menjadi numerik
mushrooms['cap-shape'] = label_encoder.fit_transform(mushrooms['cap-shape'])
mushrooms['cap-surface'] = label_encoder.fit_transform(mushrooms['cap-surface'])
mushrooms['cap-color'] = label_encoder.fit_transform(mushrooms['cap-color'])
mushrooms['bruises'] = label_encoder.fit_transform(mushrooms['bruises'])
mushrooms['odor'] = label_encoder.fit_transform(mushrooms['odor'])
mushrooms['gill-attachment'] = label_encoder.fit_transform(mushrooms['gill-attachment'])
mushrooms['gill-spacing'] = label_encoder.fit_transform(mushrooms['gill-spacing'])
mushrooms['gill-size'] = label_encoder.fit_transform(mushrooms['gill-size'])
mushrooms['gill-color'] = label_encoder.fit_transform(mushrooms['gill-color'])
mushrooms['stalk-surface-below-ring'] = label_encoder.fit_transform(mushrooms['stalk-surface-below-ring'])
mushrooms['stalk-color-above-ring'] = label_encoder.fit_transform(mushrooms['stalk-color-above-ring'])
mushrooms['stalk-color-below-ring'] = label_encoder.fit_transform(mushrooms['stalk-color-below-ring'])
mushrooms['veil-type'] = label_encoder.fit_transform(mushrooms['veil-type'])
mushrooms['veil-color'] = label_encoder.fit_transform(mushrooms['veil-color'])
mushrooms['ring-number'] = label_encoder.fit_transform(mushrooms['ring-number'])
mushrooms['ring-type'] = label_encoder.fit_transform(mushrooms['ring-type'])
mushrooms['spore-print-color'] = label_encoder.fit_transform(mushrooms['spore-print-color'])
mushrooms['population'] = label_encoder.fit_transform(mushrooms['population'])
mushrooms['habitat'] = label_encoder.fit_transform(mushrooms['habitat'])

# Menampilkan data
print(mushrooms.head())
```

```python
#memisakhan atribut dengan label dimana x itu atribut dan y itu label
X = mushrooms[['cap-color','gill-color','veil-color','spore-print-color']]
y = mushrooms['class']


#membagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

#menentukan  model decision tree
tree_model= DecisionTreeClassifier()

#melatih model menggunakan data latih dengan fungsi fit
tree_model= tree_model.fit(X_train,y_train)
```

```python
#evaluasi model untuk meguji seberapa baik model yang digunakan dan akurasi

from sklearn.metrics import accuracy_score

y_pred =  tree_model.predict(X_test)

accuracy_score =round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ',accuracy_score)
```

```python
#prediksi model dengna tree model yang sdh dibuat dengan varibel (['SepalLengthCm','SepalWidthCm','PetalWidthCm'])

print(tree_model.predict([[6, 4, 1,3]])[0])
```
